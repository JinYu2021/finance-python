```python
# 记录某项目四个季度的收入和成本（单位：万元）
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [120, 150, 180, 160]
cost = [80, 90, 110, 100]

print("数据加载成功！")
```

```python
# 计算每个季度的利润：收入 - 成本
profit = []
for i in range(len(quarters)):
    p = revenue[i] - cost[i]
    profit.append(p)

# 在 Jupyter 里，直接把变量名写在最后一行，它会自动打印出内容
profit
```

### 利润分析初步结论
经过初步核算，**第三季度 (Q3)** 的利润表现最好，达到了 70 万元。
接下来我们将对其进行可视化展示。

```python
import matplotlib.pyplot as plt

# 绘制利润柱状图
plt.bar(quarters, profit, color='skyblue')
plt.title('Quarterly Profit')
plt.ylabel('Profit')
plt.show()
```
